#### Definición de los 5 atributos
Clase: Reestructuracion_Deuda
1. nombre_cliente

Guarda el nombre de la persona que solicita la reestructuración de deuda.

2. deuda_actual

Representa la cantidad de dinero que debe el cliente actualmente.

3. tipo_deuda

Indica el tipo de deuda que tiene el cliente:

tarjeta
préstamo
hipotecaria
4. plazo_pago

Muestra los meses o años que tendrá el cliente para pagar la deuda reestructurada.

5. historial_crediticio
Indica si el historial del cliente es:
bueno
regular
malo
#### Definición de las 5 funciones ⚙️
1. agregar_cliente()

Permite registrar un nuevo cliente con sus datos financieros.

2. listar_clientes()

Muestra todos los clientes registrados en forma de tabla organizada.

3. actualizar_cliente()

Permite modificar la información de un cliente existente.

4. eliminar_cliente()

Elimina un cliente registrado en el sistema.

5. recomendacion_ia()

La inteligencia artificial recomienda opciones de pago según el historial crediticio y tipo de deuda.

```python
# ==========================================
# PROYECTO UIII ACT 13
# IA REESTRUCTURACIÓN DE DEUDA
# ==========================================

print("======================================")
print("💰 IA REESTRUCTURACIÓN DE DEUDA 💰")
print("======================================")

# CLASE
class Reestructuracion_Deuda:

    # CONSTRUCTOR
    def __init__(self):

        self.clientes = []

    # AGREGAR
    def agregar_cliente(self):

        nombre_cliente = input("Nombre del cliente: ")
        deuda_actual = input("Deuda actual: ")
        tipo_deuda = input("Tipo de deuda: ")
        plazo_pago = input("Plazo de pago: ")
        historial_crediticio = input("Historial crediticio: ")

        cliente = {

            "nombre_cliente": nombre_cliente,
            "deuda_actual": deuda_actual,
            "tipo_deuda": tipo_deuda,
            "plazo_pago": plazo_pago,
            "historial_crediticio": historial_crediticio
        }

        self.clientes.append(cliente)

        print("✅ Cliente agregado correctamente")

    # LISTAR
    def listar_clientes(self):

        if self.clientes == []:

            print("❌ No hay clientes registrados")

        else:

            print("\n============= TABLA DE CLIENTES =============")

            print("Nombre\tDeuda\tTipo\tPlazo\tHistorial")

            print("------------------------------------------------")

            for c in self.clientes:

                print(
                    c["nombre_cliente"], "\t",
                    c["deuda_actual"], "\t",
                    c["tipo_deuda"], "\t",
                    c["plazo_pago"], "\t",
                    c["historial_crediticio"]
                )

    # ACTUALIZAR
    def actualizar_cliente(self):

        nombre = input("Nombre del cliente a actualizar: ")

        encontrado = False

        for c in self.clientes:

            if c["nombre_cliente"] == nombre:

                c["deuda_actual"] = input("Nueva deuda: ")
                c["tipo_deuda"] = input("Nuevo tipo de deuda: ")
                c["plazo_pago"] = input("Nuevo plazo de pago: ")
                c["historial_crediticio"] = input("Nuevo historial: ")

                print("✅ Cliente actualizado")

                encontrado = True

        if encontrado == False:

            print("❌ Cliente no encontrado")

    # ELIMINAR
    def eliminar_cliente(self):

        nombre = input("Nombre del cliente a eliminar: ")

        encontrado = False

        for c in self.clientes:

            if c["nombre_cliente"] == nombre:

                self.clientes.remove(c)

                print("✅ Cliente eliminado")

                encontrado = True

                break

        if encontrado == False:

            print("❌ Cliente no encontrado")

    # IA
    def recomendacion_ia(self):

        print("\n🤖 RECOMENDACIÓN IA")

        historial = input("Historial crediticio: ")

        if historial == "bueno":

            print("✅ IA recomienda pagos flexibles")

        elif historial == "regular":

            print("✅ IA recomienda pagos mensuales moderados")

        elif historial == "malo":

            print("✅ IA recomienda reestructuración inmediata")

        else:

            print("❌ Historial no válido")


# OBJETO
sistema = Reestructuracion_Deuda()

# MENÚ
while True:

    print("\n=========== MENÚ ===========")
    print("1. Agregar cliente")
    print("2. Listar clientes")
    print("3. Actualizar cliente")
    print("4. Eliminar cliente")
    print("5. Recomendación IA")
    print("6. Salir")

    opcion = input("Selecciona una opción: ")

    if opcion == "1":

        sistema.agregar_cliente()

    elif opcion == "2":

        sistema.listar_clientes()

    elif opcion == "3":

        sistema.actualizar_cliente()

    elif opcion == "4":

        sistema.eliminar_cliente()

    elif opcion == "5":

        sistema.recomendacion_ia()

    elif opcion == "6":

        print("\n👋 Gracias por usar la IA Financiera")
        break

    else:

        print("❌ Opción inválida")
```
